```python
import numpy as np 
import itertools as it 
import matplotlib.pyplot as plt
```

```python
def beta(A: np.ndarray) -> float:
    d = A.shape[1]
    combinations = np.array(list(it.product([-1,1], repeat=d)))
    return np.mean(np.linalg.norm(A @ combinations.T, axis=0, ord=np.inf))
```

```python
def hadamaard_construction(n: int) -> np.ndarray:
    if n == 1:
        return np.array([[1.0, 0], [1, 1]])
    else:
        A = hadamaard_construction(n-1)
        return np.block([[A, A], [A, -A]])
```

```python
def beta_n(n: int) -> None:
    A = hadamaard_construction(n)
    A /= np.linalg.norm(A, ord = 2, axis = 1)
    print(f"n = {n}, beta = {beta(A)}")
```

```python
beta_n(4)
```

```python
def approx_beta(M: np.ndarray):
    rng = np.random.default_rng()
    X = rng.choice([-1,1], size=(len(M), 1000), p=[1/2, 1/2])
    return np.mean(np.linalg.norm(M @ X, axis=0, ord=np.inf))
```

```python
def approx_beta_n(n: int) -> None:
    A = hadamaard_construction(n)
    A /= np.linalg.norm(A, ord = 2, axis = 1)
    print(f"n = {n}, approx_beta = {approx_beta(A)}")
```

```python
for i in range(1, 10):
    matrix = hadamaard_construction(i)
    matrix /= np.linalg.norm(matrix, ord = 2, axis = 1)
    print(f"approximate beta value for", matrix.shape, "is", approx_beta(matrix))
```
